# üß© Theoretical Questions ‚Äì Clustering

---

### 1Ô∏è What is Unsupervised Learning?
   - Unsupervised learning is a branch of machine learning where we work with **unlabeled data**.  
   - The model tries to find **hidden patterns**, **relationships**, or **group structures** on its own without predefined outputs.  
  Examples include **clustering** and **dimensionality reduction**.

---

### 2Ô∏è How does the K-Means algorithm work?
K-Means divides data into **K groups (clusters)** based on feature similarity.  
The steps are:
1. Randomly choose K centroids.
2. Assign each point to the nearest centroid.
3. Recalculate centroids based on assigned points.
4. Repeat until centroids stabilize (no significant change).

---

### 3Ô∏è What is a Dendrogram in Hierarchical Clustering?
A dendrogram is a **tree-like diagram** that shows how clusters are merged or split step by step.  
It helps in visualizing the **hierarchical relationship** among clusters.

---

### 4Ô∏è Difference between K-Means and Hierarchical Clustering
- **K-Means:** Requires the number of clusters (K) beforehand.  
- **Hierarchical:** Builds a cluster tree and doesn‚Äôt need predefined K initially.

---

### 5Ô∏è Advantages of DBSCAN over K-Means
- Works well with **arbitrarily shaped clusters**.  
- **Automatically detects noise/outliers**.  
- Doesn‚Äôt need the number of clusters (K) as input.

---

### 6Ô∏è When is Silhouette Score used?
It measures **how well-separated and cohesive** the clusters are.  
Higher score ‚Üí better clustering.

---

### 7Ô∏è Limitations of Hierarchical Clustering
- Computationally expensive for large datasets.  
- Sensitive to noise and outliers.  
- Once merged, clusters cannot be undone.

---

### 8Ô∏è Importance of Feature Scaling in K-Means
Since K-Means uses **Euclidean distance**, unscaled features can dominate results.  
Scaling ensures all features contribute equally.

---

### 9Ô∏è How does DBSCAN identify noise points?
Points having **fewer than `min_samples`** within distance **`eps`** are labeled as **noise (-1)**.

---

### 10 What is Inertia in K-Means?
Inertia is the **sum of squared distances** between data points and their cluster centers.  
Lower inertia means tighter and better clusters.

---

### 11Ô∏è What is the Elbow Method?
A visual technique to determine the **optimal number of clusters (K)** by plotting **inertia vs K**.  
The ‚Äúelbow‚Äù point indicates the best K.

---

### 12Ô∏è What is ‚ÄúDensity‚Äù in DBSCAN?
Density represents the **number of points** within a given radius (`eps`).  
Dense regions form clusters; sparse regions are considered noise.

---

### 13Ô∏è Can Hierarchical Clustering be used on Categorical Data?
Yes, but we must use suitable distance measures (like **Hamming**) or encode categories numerically.

---

### 14Ô∏è What does a Negative Silhouette Score mean?
It suggests that the data point **might be in the wrong cluster**, i.e., clusters overlap or are poorly defined.

---

### 15Ô∏è Explain ‚ÄúLinkage Criteria‚Äù in Hierarchical Clustering
Determines how the **distance between clusters** is calculated:
- **Single linkage:** minimum distance  
- **Complete linkage:** maximum distance  
- **Average linkage:** mean distance

---

### 16Ô∏è Why might K-Means perform poorly on varying cluster sizes or densities?
Because K-Means assumes **spherical and equal-sized clusters**, it struggles when clusters differ in density or size.

---

### 17Ô∏è Core Parameters of DBSCAN
- `eps`: Radius of the neighborhood.  
- `min_samples`: Minimum number of points to form a dense region.  
These control cluster density and sensitivity to noise.

---

### 18Ô∏è How does K-Means++ improve initialization?
It spreads out the initial centroids more effectively, reducing chances of poor convergence and improving accuracy.

---

### 19Ô∏è What is Agglomerative Clustering?
It‚Äôs a **bottom-up** approach where each data point starts as its own cluster, and pairs of clusters are merged based on proximity until one big cluster remains.

---

### 20Ô∏è Why is Silhouette Score better than Inertia?
Silhouette Score evaluates **both cluster compactness and separation**, while inertia only measures compactness.  
Hence, Silhouette gives a more balanced view of clustering quality.

```python
# Q1. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

X, y = make_blobs(n_samples=500, centers=4, random_state=42)
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis', s=40)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, c='red', marker='X')
plt.title("K-Means Clustering with 4 Centers")
plt.show()
```

```python
# Q9. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=400, centers=4, random_state=42)
scores = []
for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    scores.append(silhouette_score(X, labels))

plt.bar(range(2,6), scores, color='teal')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for K = 2 to 5")
plt.show()
```

```python
# Q10. Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage.
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

X = load_iris().data
link = linkage(X, method='average')

plt.figure(figsize=(10,5))
dendrogram(link, leaf_rotation=90)
plt.title("Hierarchical Clustering Dendrogram")
plt.show()
```

```python
# Q11. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=400, centers=3, cluster_std=2.0, random_state=42)
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, s=40, cmap='coolwarm')
plt.title("K-Means Decision Boundaries (Overlapping Clusters)")
plt.show()
```

```python
# Q14. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot.
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

data = load_breast_cancer()
scaler = StandardScaler()
X = scaler.fit_transform(data.data)

inertia = []
for k in range(2, 7):
    km = KMeans(n_clusters=k, random_state=42).fit(X)
    inertia.append(km.inertia_)

plt.plot(range(2,7), inertia, marker='o', color='darkblue')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("K-Means Inertia Comparison (K=2 to 6)")
plt.show()
```

```python
# Q26. Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features).
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import pandas as pd
import seaborn as sns

wine = load_wine()
scaler = StandardScaler()
X = scaler.fit_transform(wine.data[:, :4])
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)
df = pd.DataFrame(wine.data[:, :4], columns=wine.feature_names[:4])
df["Cluster"] = kmeans.labels_

sns.pairplot(df, hue="Cluster", palette="husl")
plt.suptitle("KMeans Clustering on Wine Dataset (First 4 Features)", y=1.02)
plt.show()
```
